# Building our superhero by finding the right paramaters using Ai Linear regression & GD, with optimization method

# Importing libraries & Data generated by our game

```python
import pandas as pd
from scipy.optimize import minimize,Bounds,NonlinearConstraint
from math import sqrt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.metrics import mean_squared_error
df=pd.read_csv("data.csv")
df.drop(columns='type',inplace=True)
X=df.drop(columns='maxdays')
y=df['maxdays']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

# Trainging & predecting

```python

reg = LinearRegression().fit(X_train, y_train)
reg.score(X_train, y_train)

rmse = sqrt(mean_squared_error(y_test, reg.predict(X_test)))
```

# Optimization to find the max score paramaters with constraints

```python
def eq(x):
    return float((np.dot(reg.coef_,np.mat(x).T) + reg.intercept_)*-1)
def constr(x):
    return eq(x)+100
```

```python
x0 = np.array([1.3, 0.7, 0.8, 1.9])
res = minimize(eq,x0, method='nelder-mead',
                options={'xatol': 1e-8, 'disp': True},constraints=NonlinearConstraint(constr,-np.inf,100),bounds=Bounds([15,-99,0.001,0],[99,13,8000,13]))
```
